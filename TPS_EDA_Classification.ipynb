Neighbourhood and coordinate information (latitude and longitude) will appear to be
Not Specified Area (NSA) and (0,0), respectively, if any of the following conditions are met: (1)
Division is NSA OR (2) Originating X/Y values are 0 OR (3) Originating X/Y values are outside the
City of Toronto.

If an event occurred within 5,000 meters outside the City of Toronto, it is snapped to an intersection and will have
coordinates. Neighbourhood values for these events would be NSA.

June 2018 marked the amalgamation
of divisions 54 and 55 and thus after this point all offences/crimes occurring in the boundaries
of “54 Division” have been marked as “55 Division”. Please note, data summarized in the open
analytics combines all data for 54 and 55 divisions together for historical comparisons.

```python
import time
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split
```

```python
# Initialize lists to store metrics
recall_RFC = []
training_times_RFC = []
best_recall = 0
best_n_estimators = None
best_y_pred = None
time_RFC = 0

# Evaluate Random Forest for each number of estimators
for n_estimators_RFC in range(250,300, 10):
    start_time_RFC = time.time()  # Start time measurement
    rf = RandomForestClassifier(n_estimators=n_estimators_RFC, random_state=92, n_jobs=-1)
    rf.fit(X_train, y_train)
    end_time_RFC = time.time()  # End time measurement
    y_pred_RFC = rf.predict(X_test)
    recall_RFC_value = recall_score(y_test, y_pred_RFC, average='weighted')
        
    training_time_RFC = end_time_RFC - start_time_RFC  # Compute training time
    recall_RFC.append(recall_RFC_value*100)
    training_times_RFC.append(training_time_RFC)  # Append training time
    
    # Update best model if current accuracy is higher
    if recall_RFC_value  > best_recall:
        best_recall = recall_RFC_value
        best_n_estimators = n_estimators_RFC
        best_y_pred = y_pred_RFC
        time_RFC = training_time_RFC
```

```python
# Plot results
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1) # Performance Metrics  vs. number of estimators
plt.plot(range(250, 300, 10), recall_RFC, marker='o', linestyle='-', color='g',label ="Recall" )
plt.legend()
plt.xlabel('Number of Estimators')
plt.ylabel('proportions (%)')
plt.title('Recall % vs. Number of Estimators')
#training time vs. number of estimators
plt.subplot(1, 2, 2)
plt.plot(range(250, 300, 10), training_times_RFC, marker='o', linestyle='-', color='r')
plt.title('Random Forest Training Time vs. Number of Estimators')
plt.xlabel('Number of Estimators')
plt.ylabel('Training Time (seconds)')
plt.xticks(range(250, 300, 10)) 
plt.grid(True)


plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
range_k = list(range(1, 50))

# List to store the error rate values for each k
error_rate = []

for k in range_k:
    knn = KNeighborsClassifier(n_neighbors=k,n_jobs=-1)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    error_rate.append(1 - accuracy_score(y_test, y_pred))

```

```python
# Plot the error rate values to find the elbow point
plt.figure(figsize=(10, 7))
plt.plot(range_k, error_rate, marker='o', linestyle='-', color='b')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Error Rate')
plt.title('Elbow Method For Optimal k in KNN')
plt.show()
```
